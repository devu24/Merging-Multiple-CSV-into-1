```python
import pandas as pd
import glob
import os
```

```python
def merge_csv_files(folder_path, output_file):
    # Ensure the path is correctly formatted (handle spaces)
    folder_path = folder_path.strip("'")  # Remove single quotes if present
    folder_path = folder_path.replace("\\", "/")  # Normalize path for compatibility

    # Get all CSV files in the folder
    all_files = glob.glob(os.path.join(folder_path, "*.csv"))
    print(all_files)
    if not all_files:
        print("No CSV files found in the folder.")
        return

    # Read and merge all CSV files
    df_list = [pd.read_csv(file) for file in all_files]
    print(df_list)
    merged_df = pd.concat(df_list, ignore_index=True)

    # Save the merged file
    merged_df.to_csv(output_file, index=False)
    print(f"Successfully merged {len(all_files)} files into {output_file}")
```

```python
# Usage
folder_path = r"C:\Users\rahul\Downloads\Sample Dataset"  # Use raw string (r"") for Windows paths
output_file = r"C:\Users\rahul\Downloads\Sample Dataset\merged_output.csv"
merge_csv_files(folder_path, output_file)
```
